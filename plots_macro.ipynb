```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_receitas = pd.read_csv('/home/dudunene/Área de trabalho/variacoes_receita.csv').drop('Unnamed: 0', axis=1)
df_despesas = pd.read_csv('/home/dudunene/Área de trabalho/variacao_despesas.csv').drop('Unnamed: 0', axis=1)
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(20,10))

sns.barplot(data=df_receitas, x='Trimestre',y='Variacao', hue='Empresa', palette='Set2', alpha=0.9)

plt.axhline(y=1, alpha=0.9, color='black')

plt.legend(loc='upper left', fontsize=12)

plt.xlabel('Trimestre',fontsize=20)

plt.ylabel('Variacao percentual da receita', fontsize=20)

plt.title('Variacao percentual da receita de 2020 em relacao a receita media', fontsize=24, fontweight=16, pad=15)

plt.show()
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(20,10))

sns.barplot(data=df_despesas, x='Trimestre',y='Variacao', hue='Empresa', palette='Set2', alpha=0.9)

plt.axhline(y=1, alpha=0.9, color='black')

plt.legend(loc='upper left', fontsize=12)

plt.xlabel('Trimestre',fontsize=20)

plt.ylabel('Variacao percentual das despesas com venda', fontsize=20)

plt.title('Variacao percentual das despesas com venda de 2020 em relacao as despesas com venda medias', fontsize=24, fontweight=16, pad=15)

plt.show()
```
